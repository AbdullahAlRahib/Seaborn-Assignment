# **Que1**

**Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

**Answer:**


 * Scatter Plot (seaborn.scatterplot):

    Uses: Scatter plots are used to visualize the relationship between two continuous variables. They help identify patterns, correlations, and clusters in the data.

 * Line Plot (seaborn.lineplot):

    Uses: Line plots are useful for displaying trends or changes in data over a continuous variable, such as time. They are often used to visualize time series data or show the progression of values.

 * Bar Plot (seaborn.barplot):

    Uses: Bar plots are commonly used to compare and display the values of different categories. They are effective in showing categorical data and making comparisons between groups or variables.

 * Histogram (seaborn.histplot):

    Uses: Histograms are used to visualize the distribution of a single numeric variable. They display the frequency or density of values within different bins, helping to understand the underlying data distribution.

 * Heatmap (seaborn.heatmap):

    Uses: Heatmaps are used to visualize the relationship between two categorical variables. They represent data as a grid of colored cells, where each cell's color represents the value of the relationship between the variables.


# **Que2**

Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.


**Answer:**


```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
fmri_dataset = sns.load_dataset("fmri")
```

```python
sns.lineplot(data=fmri_dataset, x="timepoint", y="signal", hue="event", style="region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal by Timepoint, Event, and Region")
plt.show()
```

# **Que3**

Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

    
**Answer:**


```python
titanic_dataset = sns.load_dataset("titanic")
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_dataset, x='pclass', y='age')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.title('Box Plot: Age by Passenger Class')


plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_dataset, x='pclass', y='fare')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Box Plot: Fare by Passenger Class')
plt.tight_layout()
plt.show()
```

# **Que4**

**Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.**

**Answer:**



```python
diamonds_dataset = sns.load_dataset("diamonds")
```

```python
sns.histplot(data=diamonds_dataset, x='price', hue='cut', kde=True)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Diamond Prices by Cut')
plt.show()
```

# **Que6**

**6: Use the "flights" dataset from seaborn to plot a heatmap.**

**Answer:**


```python
flights_dataset = sns.load_dataset("flights")
```

```python
flights_matrix = flights_dataset.pivot("month", "year", "passengers")

# Plot the heatmap
sns.heatmap(data=flights_matrix, cmap="YlGnBu")

# Set labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Flights Heatmap")

# Display the plot
plt.show()
```
